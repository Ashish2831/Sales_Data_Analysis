# Task7

```python
# Finding out the most and least sold product id
# Create a new dataset including the product ids and total quantities sold for that id

mostAndLeast = pd.DataFrame(df['Product ID'])
mostAndLeast['Product Name'] = df['Product Name']
mostAndLeast['Quantity'] = df['Quantity']
grpmostAndLeast = mostAndLeast.groupby('Product ID', as_index=False).sum()
grpmostAndLeast
```

```python
LeastQuantity = grpmostAndLeast['Quantity'].min()
LeastQuantity
```

```python
MostQuantity = grpmostAndLeast['Quantity'].max()
MostQuantity
```

```python
# Find out the least sold product from the created dataset

LeastProductsSold = grpmostAndLeast[grpmostAndLeast['Quantity'] == LeastQuantity]
LeastProductsSold
```

```python
# Find out the most sold product from the created dataset

MostProductsSold = grpmostAndLeast[grpmostAndLeast['Quantity'] == MostQuantity]
MostProductsSold.set_index(pd.Index(['MostProductSold']))
```

```python
productIDInfo = df[df['Product ID'].str[0:] == 'TEC-CO-10004722']
productName = pd.DataFrame(productIDInfo['Product Name'])
productName = productName.drop_duplicates()
productName.set_index(pd.Index(['MostProductSold']))
```
